```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("company_esg_financial_dataset.csv")
```

```python
df['GrowthRate'].fillna(df['GrowthRate'].mean(), inplace=True)
```

```python
df.isnull().sum()
```

```python
import pandas as pd

df['ESG_Category'] = pd.cut(
    df['ESG_Overall'],
    bins=[0, 50, 75, 100],
    labels=['Low', 'Medium', 'High']
)
```

```python
df['Industry'].unique()
df['Region'].unique()
```

```python
colors = ['#d62728', '#ff7f0e', '#2ca02c']  # Red, Orange, Green

df['ESG_Category'].value_counts().plot.pie(
    autopct='%1.1f%%',
    startangle=90,
    colors=colors,
    title='ESG Category Share'
)
plt.ylabel('')  
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

pivot = df.pivot_table(
    index='Region', 
    columns='Industry', 
    values='ESG_Overall', 
    aggfunc='mean'
).fillna(0)

plt.figure(figsize=(12,6))
sns.heatmap(pivot, annot=True, fmt=".1f", cmap='YlGnBu')
plt.title('Average ESG Score by Region and Industry')
plt.show()
```

```python
df['CarbonIntensity'] = df['CarbonEmissions'] / df['Revenue']

region_carbon = df.groupby('Region')['CarbonIntensity'].mean().reset_index()
print(region_carbon.sort_values(by='CarbonIntensity', ascending=False))
```

```python
carbon_trend = df.groupby('Year')['CarbonIntensity'].mean()
carbon_trend.plot(title='Average Carbon Intensity Over Years', marker='o')
plt.ylabel('CO per Revenue Unit')
plt.show()
```

```python
df_sorted = df.sort_values(['CompanyID', 'Year'])
df_sorted['ESG_Rolling3'] = df_sorted.groupby('CompanyID')['ESG_Overall'].transform(lambda x: x.rolling(window=3, min_periods=1).mean())

sample = df_sorted[df_sorted['CompanyID'] == df_sorted['CompanyID'].iloc[0]]
sample.plot(x='Year', y=['ESG_Overall', 'ESG_Rolling3'], title=f"Company {sample['CompanyName'].iloc[0]} ESG Trend")
plt.show()
```

```python
esg_trend = df.groupby('Year')[['ESG_Overall', 'ESG_Environmental', 'ESG_Social', 'ESG_Governance']].mean()
esg_trend.plot(title="Average ESG Scores Over Time", figsize=(10,5));
```

```python
industry_esg = df.groupby('Industry')[['ESG_Overall', 'CarbonEmissions', 'WaterUsage']].mean().sort_values(by='ESG_Overall', ascending=False)
industry_esg.plot(kind='bar', figsize=(12,6), title="Industry-wise ESG Scores & Environmental Impact");
```

```python
top_esg = df[df['ESG_Overall'] > 85][['CompanyName', 'Year', 'Industry', 'ESG_Overall']].sort_values(by='ESG_Overall', ascending=False)
bottom_esg = df[df['ESG_Overall'] < 40][['CompanyName', 'Year', 'Industry', 'ESG_Overall']].sort_values(by='ESG_Overall')

```

```python
financial_cols = ['Revenue', 'ProfitMargin', 'MarketCap', 'GrowthRate']
esg_fin_corr = df[['ESG_Overall'] + financial_cols].corr()
print(esg_fin_corr)
sns.heatmap(esg_fin_corr, annot=True, cmap='coolwarm')
plt.title("Correlation between ESG and Financial Metrics")
plt.show()
```

```python
region_env = df.groupby('Region')[['CarbonEmissions', 'WaterUsage', 'EnergyConsumption']].mean().sort_values(by='CarbonEmissions', ascending=False)
region_env.plot(kind='bar', figsize=(10,5), title="Average Environmental Metrics by Region");
```

```python
df.to_csv("Cleaned_ESG_Analysis.csv", index=False)
top_esg.to_csv("Top_ESG_Companies.csv", index=False)
```
